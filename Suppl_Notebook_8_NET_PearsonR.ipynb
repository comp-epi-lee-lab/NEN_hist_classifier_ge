## Pearson R of specific samples 

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("gene_data_PDWD_deID.csv", header=0, index_col = 0)

df
```

```
import numpy as np

df_log = np.log( df+1 )
df_log
```

```
df_log.loc[:,['WD_0021', 'WD_0012']]

```

```
from scipy import stats

stats.pearsonr(df_log.loc[:,'WD_0012'], df_log.loc[:,'WD_0012'])
```

```
[r, p] = stats.pearsonr(df_log.loc[:,'WD_0012'].tolist(), df_log.loc[:,'WD_0009'].tolist() )
print ("Pearson correlation r = {:.2f} (p-val={:.2e})".format(r, p))
```

```
df_log.loc[:,['PD_0015', 'PD_0013']]
[r, p] = stats.pearsonr(df_log.loc[:,'PD_0015'].tolist(), df_log.loc[:,'PD_0013'].tolist() )
print ("Pearson correlation r = {:.2f} (p-val={:.2e})".format(r, p))
```

```
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))
plt.scatter(
    df_log["WD_0021"],
    df_log["WD_0009"],
    color="blue",
    alpha=0.2,
    edgecolors="none",
)

plt.xlabel("WD_0021", fontsize=14)
plt.ylabel("WD_0009", fontsize=14)
plt.title("")
plt.grid(False)

plt.xticks(range(0, 12, 1), fontsize=14)  
plt.yticks(range(1, 12, 1), fontsize=14)  


r, p = stats.pearsonr(df_log["WD_0021"], df_log["WD_0009"])
correlation_text = "Pearson correlation r = {:.2f}".format(r) + "\n(p-val={:.2e})".format(p)
plt.text(0.05, 0.95, correlation_text, transform=plt.gca().transAxes, va="top", fontsize=14)

# Show the plot

plt.savefig("WD_0021_WD_0009_Correlation_r.pdf")

plt.show()

```
